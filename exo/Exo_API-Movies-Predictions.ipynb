```python
import pandas as pd
import time
from bs4 import BeautifulSoup
import requests
import json
```

## Scrapping de la page des films en random

```python
raw_html = requests.get('https://www.listchallenges.com/completely-random-movie-list').text
raw_html
```

```python
html = BeautifulSoup(raw_html, 'lxml') 
html
```

#### Get Movies list

```python
item_names = html.select('div[class="item-name"]')
item_names
```

```python
movies_list = [
    item_name.get_text(strip = True) 
    for item_name in item_names
]

movies_list
```

#### Nettoyage des titres de film

```python
movies = [movie.split(" (", )[0] for movie in movies_list]
movies
```

## Récupération des métadata des films
Site: http://www.omdbapi.com/  
API call: http://www.omdbapi.com/?apikey=[yourkey]&  
API Key: 7e4266a4

```python
df = pd.DataFrame(
    [
        requests.get('http://www.omdbapi.com/?apikey=7e4266a4&t={}'.format(movie)).json() 
        for movie in movies
    ]
)

df
```

## Algo machine Learning
### Détermination du IMDB score du film via regression Linéaire
#### Nettoaye du Dataset

```python
df['BoxOffice'] = df['BoxOffice'].apply(
    lambda x: x.replace('$','').replace(',','') 
    if x!='N/A' 
    else None
)
```
